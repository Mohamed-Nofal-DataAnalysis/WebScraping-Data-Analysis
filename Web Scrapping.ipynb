```python
from bs4 import BeautifulSoup
import requests 
import pandas as pd 
import csv
from itertools import zip_longest
```

```python
url='https://wuzzuf.net/search/jobs?q=data%20anlysis'
page=requests.get(url)
source=page.content
source
```

```python
scoup=BeautifulSoup(source,'lxml')
scoup
```

```python
job_titles=scoup.find_all("h2",{"class":"css-193uk2c"})
company_names=scoup.find_all("a",{"class":"css-ipsyv7"})
locations=scoup.find_all("span",{"class":"css-16x61xq"})
job_types=scoup.find_all("div",{"class":"css-5jhz9n"})
len(job_titles)
```

```python
initial_divs = scoup.find_all("div", {"class": "css-5jhz9n"})
skill = []
for div in initial_divs :
    skill_div=div.find_next_sibling('div')
    if skill_div :
        skill.append(div.text.strip())
    else :
        skill.append('Not Found')
```

```python
job_title= []
company_name= []
location= []
job_type= []
links=[]
```

```python
for i in range(len(job_titles)):
    job_title.append(job_titles[i].text.strip())
    company_name.append(company_names[i].text.strip())
    location.append(locations[i].text.strip())
    job_type.append(job_types[i].text.strip())
    links.append(job_titles[i].find("a").attrs['href'])
```

```python
# This is critical for data alignment.
print(f"Job Titles: {len(job_title)}")
print(f"Company Names: {len(company_name)}")
print(f"Locations: {len(location)}")
print(f"Skills: {len(skill)}")

# Proceed only if the lengths match
if len(job_title)==len(company_name)==len(location)==len(skill):
    # Create a DataFrame
    jobs_df = pd.DataFrame({
        'Job Title': job_title,
        'Company Name': company_name,
        #'Experience Need' : experience_need,
        'Location': location,
        'Skills': skill,
        'Job Type': job_type,
        'Links': links
    })

    # Display the DataFrame
    print(jobs_df.head())
```

```python
jobs_df.to_excel('jobs_data.xlsx', index=False)
```
